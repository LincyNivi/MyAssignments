```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
dataset=pd.read_csv('/content/Market_Basket_Optimisation.csv',header=None)
```

```python
dataset.head(10)
```

```python
dataset.shape  ## 7501 transactions
```

```python
# For using Aprori , need to convert data in list format
transactions=[]

for i in range(0,7501):
    transactions.append([str(dataset.values[i,j])
                         for j in range(0,20)])
```

```python
transactions[700]
```

```python
# Training the Apriori model on the dataset

Apriori_Rules= apriori(transactions=transactions,  ## dataset
                      min_support= 0.001,  ## 0.003
                      min_confidence=0.003,
                      min_lift=3,
                      min_length=2,   ##
                      max_length=2)   ##
```

```python
Results=list(Apriori_Rules)
```

```python
Results
```

```python
# Putting the Results in a well organised way using Padas Datafrmae
def inspect(Results):
    lhs=[tuple(result[2][0][0])[0] for result in Results]
    rhs=[tuple(result[2][0][1])[0] for result in Results]
    support=[result[1] for result in Results]
    confidences=[result[2][0][2] for result in Results]
    lifts= [result[2][0][3] for result in Results]
    return list(zip(lhs,rhs,support,confidences,lifts))
dataframe_final=pd.DataFrame(inspect(Results),columns=['Left Hand Side','Right Hand Side','Support','Confidence',
                                                    'Lift'])
```
